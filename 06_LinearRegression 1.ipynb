```python
import torch
```

```python
w = torch.tensor(3.0, requires_grad=True)
b = torch.tensor(1.0, requires_grad=True)
```

```python
def forward(x):
  y = w * x + b
  return y
```

```python
x = torch.tensor(2)
forward(x)
```

```python
x = torch.tensor([[4], [7]])
forward(x)
```
