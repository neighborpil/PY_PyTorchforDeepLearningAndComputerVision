```python
import torch
import numpy as np
```

```python
v = torch.tensor([1, 2, 3])
print(v)
print(v.dtype)
print(v[0])
print(v[1])
print(v[2])
```

```python
v = torch.tensor([1, 2, 3, 4, 5, 6])
print(v[1:4])
print(v[1:])
```

```python
f = torch.FloatTensor([1, 2, 3, 4, 5, 6])
print(f)
print(f.dtype)
print(f.size)
```

```python
v.view(6, 1) # reshpage the tensor. similar to numpy reshape
```

```python
v.view(3, 2)
```

```python
v.view(3, -1) # 마지막에 -1을 해주면 마지막 행렬의 수를 자동으로 계산해 준다
```

```python
a = np.array([1, 2, 3, 4, 5])
tensor_cnv = torch.from_numpy(a)
print(tensor_cnv)
print(tensor_cnv.type())
```

```python
numpy_cnv = tensor_cnv.numpy()
print(numpy_cnv)
```
